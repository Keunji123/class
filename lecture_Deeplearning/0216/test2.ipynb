```python
import pymysql
from easydict import EasyDict

settings = EasyDict()

settings.DB_HOST = ""
settings.DB_USER = ""
settings.DB_PASSWORD = ""
settings.DB_NAME = ""
settings.DB_PORT = 
```

```python
db = pymysql.connect(
    host = settings.DB_HOST,
    user = settings.DB_USER,
    passwd = settings.DB_PASSWORD,
    db = settings.DB_NAME,
    port = settings.DB_PORT
)
cursor = db.cursor()
cursor
```

```python
sql = 'SELECT * FROM hotel LIMIT 10;'

result = cursor.execute(sql)
result
```

```python
data = cursor.fetchall()
data
```

```python
for elem in data:
    print(elem)
```

```python
cursor.description
```

```python
columns = [elem[0] for elem in cursor.description]
```

```python
import pandas as pd

df = pd.DataFrame(data, columns = columns)
```

```python
df
```

```python
def get_query(sql: str):
    cursor.execute(sql)
    
    data = cursor.fetchall()
    columns = [elem[0] for elem in cursor.description]
    
    return pd.DataFrame(data, columns = columns)
```

```python
sql = 'SELECT * FROM hotel LIMIT 12;'

df = get_query(sql)
df
```

```python
sql = "SELECT * FROM hotel WHERE arrival_date_year = 2015 && arrival_date_month = 'July';"

df = get_query(sql)
df
```
